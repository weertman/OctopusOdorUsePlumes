```python
import os
import glob
import shutil
import pandas as pd
```

```python
target_dir = '../../DATA/1T/trajectories/'
if os.path.exists(target_dir):
    print('Directory already exists')
else:
    os.mkdir(target_dir)
    print('Directory created')
```

```python
animals = ['khorne', 'korra', 'kratos', 'larsson', 'ninja']
sub_dirs = ['No_food_Control', 'Food_eaten', 'Food_not_eaten']

for animal in animals:
    print(animal)
    animal_dir = os.path.join(target_dir, animal)
    if os.path.exists(animal_dir):
        print(animal_dir, '\tDirectory already exists')
    else:
        os.mkdir(animal_dir)
        print(animal_dir, '\tDirectory created')

    for sub_dir in sub_dirs:
        sub_dir = os.path.join(animal_dir, sub_dir)
        if os.path.exists(sub_dir):
            print(sub_dir, '\t\tDirectory already exists')
        else:
            os.mkdir(sub_dir)
            print(sub_dir, '\t\tDirectory created')
```

```python
root_controls = '../../DATA/sinlge_target_experiment_trajs__400_220_1550_965'
if os.path.exists(root_controls) != True:
    print('Directory does not exist')
```

```python
def get_dir_num (sub_dir):
    return int(os.path.basename(sub_dir).split('_')[0])

for animal in animals:
    print(animal)

    sub_dirs = glob.glob(os.path.join(root_controls, animal, '*'))
    sub_dirs = [s for s in sub_dirs if os.path.isdir(s)]
    sub_dirs = sorted(sub_dirs, key=get_dir_num)

    no_food_controls_dirs = []
    food_not_eaten_dirs = []

    for i, sub_dir in enumerate(sub_dirs):
        print('\t', sub_dir)

        path_food_annotation = os.path.join(sub_dir, 'food_annotation.csv')

        food_annotation = pd.read_csv(path_food_annotation)['NoFood0-BackFoodE1-BackFoodNE2-OtherFood3-Dirty4'].values[0]
        if food_annotation == 0:
            no_food_controls_dirs.append(sub_dir)
        elif food_annotation == 2:
            food_not_eaten_dirs.append(sub_dir)
        else:
            print('pass')
            continue

    for i, sub_dir in enumerate(no_food_controls_dirs):

        sub_dir_target = os.path.join(target_dir, animal, 'No_food_Control', str(i))
        if os.path.exists(sub_dir_target):
            print(sub_dir_target, '\t\tDirectory already exists')
        else:
            os.mkdir(sub_dir_target)
            print(sub_dir_target, '\t\tDirectory created')

        path_background_sub_ff_video = os.path.join(sub_dir, 'background_sub_ff_video.mp4')
        path_ff_video = glob.glob(os.path.join(sub_dir, '*_ff.mp4'))[0]
        path_tracking_data = glob.glob(os.path.join(sub_dir, '*_ff.h5'))[0]

        files_to_be_copied = [path_background_sub_ff_video, path_ff_video, path_tracking_data]

        for file in files_to_be_copied:
            file_target = os.path.join(sub_dir_target, os.path.basename(file))
            if os.path.exists(file_target) != True:
                shutil.copyfile(file, file_target)
                print('\t\t\t', file_target, '\t\tFile copied')
            else:
                print('\t\t\t', file_target, '\t\tFile already exists')

    for i, sub_dir in enumerate(food_not_eaten_dirs):

        sub_dir_target = os.path.join(target_dir, animal, 'Food_not_eaten', str(i))
        if os.path.exists(sub_dir_target):
            print(sub_dir_target, '\t\tDirectory already exists')
        else:
            os.mkdir(sub_dir_target)
            print(sub_dir_target, '\t\tDirectory created')

        path_background_sub_ff_video = os.path.join(sub_dir, 'background_sub_ff_video.mp4')
        path_ff_video = glob.glob(os.path.join(sub_dir, '*_ff.mp4'))[0]
        path_tracking_data = glob.glob(os.path.join(sub_dir, '*_ff.h5'))[0]

        files_to_be_copied = [path_background_sub_ff_video, path_ff_video, path_tracking_data]

        for file in files_to_be_copied:
            file_target = os.path.join(sub_dir_target, os.path.basename(file))
            if os.path.exists(file_target) != True:
                shutil.copyfile(file, file_target)
                print('\t\t\t', file_target, '\t\tFile copied')
            else:
                print('\t\t\t', file_target, '\t\tFile already exists')


```
