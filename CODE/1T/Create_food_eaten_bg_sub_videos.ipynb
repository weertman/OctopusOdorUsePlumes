```python
import os
import glob
import numpy as np
from tqdm import tqdm
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
root_dir = '../../DATA/1T/trajectories'
if os.path.exists(root_dir):
    print('Root directory exists')
else:
    print('Root directory does not exist')

food_head = 'Food_eaten'
```

```python
animals = ['khorne', 'korra', 'kratos', 'larsson', 'ninja']

video_lookup = {}

for animal in animals:
    print(animal)

    food_dir = os.path.join(root_dir, animal, food_head)
    print('\t', food_dir)

    sub_dirs = glob.glob(os.path.join(food_dir, '*'))

    for sub_dir in sub_dirs:
        path_video = glob.glob(os.path.join(sub_dir, '*_ff.mp4'))[0]
        path_data = glob.glob(os.path.join(sub_dir, '*.h5'))[0]
        print('\t\t', path_video)
        print('\t\t\t', path_data)
        video_lookup[path_video] = path_data
```

```python

pbar = tqdm(total = len(video_lookup.keys()), position=0, leave=True)
for i, path_video in enumerate(video_lookup.keys()):
    path_bg_sub_video = os.path.join(os.path.dirname(path_video), 'background_sub_ff_video.mp4')
    if os.path.exists(path_bg_sub_video) != True:
        path_data = video_lookup[path_video]
        print(i, path_video)

        data = pd.read_hdf(path_data)
        cap = cv2.VideoCapture(path_video)
        frames = []
        print('\t', 'Reading frames')
        vbar = tqdm(total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)), position=0, leave=True)
        for i in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
            ret, frame = cap.read()
            if ret:
                frames.append(frame)
            else:
                break
            vbar.update(1)
        vbar.close()
        cap.release()

        lex = data[data.columns[0]]
        ley = data[data.columns[1]]
        rex = data[data.columns[3]]
        rey = data[data.columns[4]]
        mex = (lex+rex)/2
        mey = (ley+rey)/2

        ylim = (1200, 0)
        xlim = (0, 1920)

        bbox_x = [400, 1550, 1550, 400, 400]
        bbox_y = [220, 220, 965, 965, 220]

        bbox_y_pad = 75
        bbox_x_pad = 75

        bbox_x_padded = [bbox_x[0]-bbox_x_pad, bbox_x[1]+bbox_x_pad, bbox_x[2]+bbox_x_pad, bbox_x[3]-bbox_x_pad, bbox_x[0]-bbox_x_pad]
        bbox_y_padded = [bbox_y[0]-bbox_y_pad, bbox_y[1]-bbox_y_pad, bbox_y[2]+bbox_y_pad, bbox_y[3]+bbox_y_pad, bbox_y[0]-bbox_y_pad]

        ## create true false list
        ## true if mex, mey within bbox_x_padded, bbox_y_padded
        ## false otherwise
        print('\t', 'Creating true/false list')
        inBoxIdxs = []
        inBox = []
        # Check if the mid-eye points are within the padded bounding box
        for i, (x, y) in enumerate(list(zip(mex, mey))):
            if (x >= bbox_x_padded[0] and x <= bbox_x_padded[1]) and (y >= bbox_y_padded[0] and y <= bbox_y_padded[2]):
                inBox.append(True)
            else:
                inBoxIdxs.append(i)
                inBox.append(False)
        frames_in_box = [frames[i] for i in inBoxIdxs]
        print('\t', 'Number of frames in box:', len(frames_in_box), 'out of', len(frames))
        print('\t', 'Calculating mean of frames')
        mean_of_frames = np.mean(frames_in_box, axis=0)
        print('\t', 'Calculating std of frames')
        std_of_frames = np.std(frames_in_box)

        print('\t', 'Plotting trajectory and mean')
        fig, axs = plt.subplots(1, 2, figsize=(18, 6))

        ax = axs[0]

        # Plot bounding boxes
        ax.plot(bbox_x, bbox_y, 'r-', label="Original BBox")
        ax.plot(bbox_x_padded, bbox_y_padded, 'g--', label="Padded BBox")

        # Plot the mid-eye points
        colors = ['g' if box else 'r' for box in inBox]
        for i in range(len(mex)-1):
            ax.plot(mex[i:i+2], mey[i:i+2], color=colors[i])

        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
        ax.set_aspect('equal')

        ax = axs[1]
        ax.imshow(mean_of_frames/255, cmap='gray')
        ax.set_title('Mean of frames')

        plt.show(fig)
        plt.close(fig)

        modified_frames = []
        tqdm.write('Modifying frames')
        vbar = tqdm(total = len(frames), position=0, leave=True)
        for frame in frames[:]:
            modified_frame = frame.copy()
            modified_frame = modified_frame-mean_of_frames
            modified_frame /= 255

            modified_frame[modified_frame > (5*std_of_frames)] = 1
            modified_frame[modified_frame < 0] = 0

            modified_frames.append(modified_frame)
            vbar.update(1)
        vbar.close()

        ncols = 5
        nrows = 1
        fig, axs = plt.subplots(nrows, ncols, figsize=(ncols*3, nrows*2))

        print('\t', 'Plotting some modified frames')
        step_by = int(len(modified_frames)/ncols)
        for j,frame in enumerate(modified_frames[::step_by]):
            ax = axs[j]
            ax.imshow(frame, cmap='gray')
            ax.axis('off')

        plt.show(fig)
        plt.close(fig)

        print('\t', 'Saving modified frames to video')
        print('\t', path_bg_sub_video)
        video = cv2.VideoWriter(path_bg_sub_video ,
                                cv2.VideoWriter_fourcc(*'mp4v'),
                                10,
                                (1920, 1200))
        vbar = tqdm(total = len(modified_frames), position=0, leave=True)
        for frame in modified_frames[:]:
            frame = frame*255
            frame = frame.astype(np.uint8)
            video.write(frame)
            vbar.update(1)
        vbar.close()
        video.release()

        pbar.update(1)
    else:
        print(i, path_video, 'already exists')
        pbar.update(1)

pbar.close()
```
